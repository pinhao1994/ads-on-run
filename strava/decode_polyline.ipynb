```python
import pandas as pd
import polyline
```

```python
strava = pd.read_csv("../data-collect/NYC_Strava.csv", index_col="Unnamed: 0")
strava
```

```python
lat_long = dict()
for i, row in strava.iterrows():
    lats, lons = zip(* polyline.decode(row["map_polyline"]))
    lat_long[row["name"]] = {"lats": lats, "longs": lons}
```

```python
for key, val in lat_long.items():
    pd.DataFrame(val).to_csv(f"{key}.csv")
```
